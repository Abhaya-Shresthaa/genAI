```python
from langchain.tools import BaseTool
from pydantic import BaseModel, Field
from typing import Type
```

```python
#making basic pydantic class
class MultiplyInput(BaseModel):
    a: int = Field(required = True, description="First number to multiply")
    b: int = Field(required = True, description="Second number to multiply")
```

```python
# making a tool
class MultiplyTool(BaseTool):
    name: str = 'multiply' #name to the tool
    description: str = 'Multiply two numbers' #description of tool
    args_schema: Type[BaseModel] = MultiplyInput #schema i.e pydantic strict type definition
    
    #function to use in the tool
    #name should be _run compulsory
    def _run (self, a: int, b: int) -> int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a': 2, 'b': 2})

print(result)
```
