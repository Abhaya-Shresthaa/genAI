```python
from langchain_core.tools import tool
```

toolkit is package of multiple functions or tools in single of similar kind

```python
# custom tools

@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a+b

@tool
def multiply(a:int, b: int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolKit()
```

```python
tools = toolkit.get_tools()
print(tools)
```

```python
for tool in tools:
    print(tool.name, '----->', tool.description)
```
