```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
#making basic pydantic class
class MultiplyInput(BaseModel):
    a: int = Field(required = True, description="First number to multiply")
    b: int = Field(required = True, description="Second number to multiply")
```

```python
#making function for this
def multiply_func(a: int, b: int) -> int:
    return a*b
```

```python
# actual making it tool
multiply_tool = StructuredTool.from_function(
    func= multiply_func, #function to use as tool
    name= 'multiply', #name to show for that function
    description= "Multiply 2 numbers",
    args_schema=MultiplyInput #pydantic for making strict rules in input and outputs
    
 )
```

```python
result = multiply_tool.invoke({'a': 2, 'b': 2})

print(result)
```
