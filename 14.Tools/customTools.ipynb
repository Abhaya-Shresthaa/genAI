```python
from langchain_core.tools import tool
```

```python
#create a function
#add type hints
#add tool decorator i.e @tools
#box string for tools descriptions

@tool
def multiply(a: int, b: int) -> int: #type defined here args as well as returning type
    """Multiply two numbers""" #box sting tool description
    return a*b
```

```python
result = multiply.invoke({'a': 1, 'b':2})
```

```python
print(result)
```

```python
print(multiply.name)
print()
print(multiply.description)
print()
print(multiply.args)
```
