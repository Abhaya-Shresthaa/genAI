```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
```

```python
loader = DirectoryLoader(
    path= 'directory_testing',
    glob= '*.pdf',      # import pdf files only
    loader_cls= PyPDFLoader
)
```

"**/*.txt"     for importing txt files from all subfolders as well.  
"*.pdf"        pdf files from root directory  
"data/*.csv"   importing csv files from data folder.  
"**/*"         all files any type and all folders     
**             for recursive search through subfolders

```python
docs = loader.load()
```
